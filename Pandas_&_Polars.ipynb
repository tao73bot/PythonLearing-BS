## Map And Trnasform

```python
def double_value(x):
  return x*2
```

```python
# Map is only applicable on series
pd_sales['BasketValue'].map(double_value)
```

```python
# Apply change all value of dataset
pd_sales.apply(double_value)
```

```python
# Transform takes lamda funtion
pd_sales.transform(lambda x : x*2)
```

### Mergeing and Joining

### Merging

```python
df1 = pd.DataFrame({"Key":['A','B'],"Value":[1,2]})
df2 = pd.DataFrame({"Key":['A','B','C'],"Value":[1,2,3]})
```

```python
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

### join

```python
df1.set_index("Key").join(df2.set_index("Key"),how='inner')
```
